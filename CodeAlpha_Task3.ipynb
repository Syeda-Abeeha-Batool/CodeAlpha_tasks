# Task 3 Basic ChatBot

```python
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from datetime import datetime
import difflib

def welcome_user():
    print("ChatBOT !!!!".center(100, " "))


def greetings(message):
    lower_message = message.lower()
    capital_message = message.capitalize()
    greetings = ["Hello","Hi","Hey", "Greetings","Good morning","Good afternoon","Good evening", "Sup","Hi there"]
    if any(phrase in message for phrase in greetings) or any(phrase in lower_message for phrase in greetings) or any(phrase in capital_message for phrase in greetings):
            print("BOT: Hi there! How can I help you today?")
            return True
    return False


def goodbye(message):
    lower_message = message.lower()
    capital_message = message.capitalize()
    good_bye = ["Goodbye","Bye","Farewell","See you","See you later","See you soon","See you tomorrow","Take care","Until next time","Later","Peace out","Have a great day"]
    if any(phrase in message for phrase in good_bye) or any(phrase in lower_message for phrase in good_bye) or any(phrase in capital_message for phrase in good_bye):
        print("BOT: Goodbye! Take care and have a great day!")
        return True
    return False


def mathematical_questions(message):
    addition = ["Add","Plus","Sum","+"]
    multiply = ["Multiply", "Product", "Multiplication", "Multiplied","*"]
    division = ["Divide", "Division", "Divided", "Ratio","/"]
    subtract = ["Subtract", "Minus", "Difference","-"]
    
    tokens = word_tokenize(message)
    for t in tokens:
        lower = t.lower()
        capital = t.capitalize()
        numbers = []
        for k in tokens:
            if k.isdigit():
                numbers.append(int(k))
        if len(numbers) == 2:
            if t in addition or lower in addition or capital in addition:
                print("BOT: \"The sum of ", numbers[0]," and ", numbers[1], " is : ", numbers[0] + numbers[1], "\"")
                return True
            elif t in multiply or lower in multiply or capital in multiply:
                print("BOT: \"The product of ", numbers[0]," and ", numbers[1], " is : ", numbers[0] * numbers[1], "\"")
                return True
            elif t in division or lower in division or capital in division:
                print("BOT: \"The division of ", numbers[0]," and ", numbers[1], " is : ", numbers[0]/numbers[1], "\"")
                return True
            elif t in subtract or lower in subtract or capital in subtract:
                print("BOT: \"The difference of ", numbers[0]," and ", numbers[1], " is : ", numbers[0]-numbers[1], "\"")
                return True
        else:
            return False


def response_user(message, data):
    tokens = word_tokenize(message)
    stop_words = set(stopwords.words("english"))

    for i in range(0, len(data), 2):
        matching_count = 0
        question_tokens = word_tokenize(data[i])

        found = True  
        for t in tokens:
            if t not in stop_words and t not in question_tokens:
                found = False
                break

        mess = []
        for z in tokens:
            l = z.lower()
            if l not in stop_words:
                mess.append(l)
        qq = []        
        for q in question_tokens:
            ll = q.lower()
            if q not in stop_words:
                qq.append(ll)
                
        m_string = " ".join(mess)
        q_string = " ".join(qq)
        if found or difflib.SequenceMatcher(None, m_string, q_string).ratio() > 0.6:
            if data[i + 1] == "time" or data[i + 1] == "date" or data[i + 1] == "day":
                time = datetime.now()
                print("Current Time and Date:", time)
                days_of_week = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
                time = datetime.now()
                day_of_week = time.weekday()
                current_day = days_of_week[day_of_week]
                print("Current Day of the Week:", current_day)
            else:
                print("BOT:", data[i + 1])
            return

    print("BOT: I'm sorry, I couldn't find a suitable response.")

    
def chatbot(data):
    welcome_user()
    while True:
        user = input("Enter your message...\n")
        print("USER: \"", user, "\"")
        g = greetings(user)
        if g == True:
              continue
        else:
            exit = goodbye(user)
            if exit == True:
                break
            else:
                m = mathematical_questions(user)
                if m == False:
                    response_user(user, data)
            
    
    
# ChatBot Main:    
with open("questions.txt", "r") as file:
    data = file.readlines()
    
for i in range(len(data)):
    data[i] = data[i].replace("\n", "")
    
chatbot(data)
```
