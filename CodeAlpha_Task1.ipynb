# Task 1 HangMan Game

```python
from random_word import RandomWords


print("HANGMAN GAME !!!!".center(100, " "))
print("\n\nGuess the word.....")

words = RandomWords()
random_word = words.get_random_word().lower()
random_word_list = list(random_word)

length = len(random_word)

word_list = ["_"] * length

word_string = " ".join(word_list)
print(word_string)
    
    
errors = ["x"] * 7
errors_string = " ".join(errors)
error_count = 0
print("\nWrong guesses : \n", errors_string)

while "_" in word_list and error_count < 7:
    n = input("\nEnter a letter to guess the word: ")
    
    if n in random_word:
        for k in range(length):
            if n == random_word_list[k]:
                word_list[k] = n

        word_string = " ".join(word_list)
        print(word_string)
        print("\nWrong guesses : \n", errors_string)
    else:
        errors[error_count] = n
        error_count = error_count+1
        errors_string = " ".join(errors)
        print("\nWrong guesses : \n", errors_string)
        print("\n",word_string)
        
        

if "_" not in word_list:
    print("Congratulations!!! You Guessed the Word...")
else:
    print("\nOops!!! You failed to guess the word. The correct word was '",random_word, "'")
```
