# Task 4 Data Cleaning Automation with Python Scripts

```python
import pandas as pd                 # Replace column names with your actual data columns

input_file = "input_file.csv"       # Replace with your actual file name and path
df = pd.read_csv(input_file)

df = df.dropna(subset = ["Name"])     # Drop rows with NaN values in the "Name" column

numeric_columns = ["Revenue (USD millions)", "Revenue growth (%)", "Employees"]

df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric, errors='coerce')       # Convert columns to numeric, replacing non-numeric values with NaN

df[numeric_columns] = df[numeric_columns].fillna(0)         # Replacing the NaN values with default value, 0

df = df.drop_duplicates(subset = ["Name"])        # Drop duplicate rows in the "Name" column

df.to_csv("output_file.csv", index=False)
print(df.info())
df
```
